### Encoder

To implement BERT's encoder layer for the Transformer, developers need to construct two primary sub-layers: a multi-headed attention layer (already partially implemented) and a feed-forward layer. The implementation requires developing the feed-forward sub-layer and incorporating residual connections, which include the Add & Normalize blocks as depicted in the architectural diagram. <br>

```python
class MultiHeadAttention(layers.Layer):
    def __init__(self, hidden_size, num_heads):

        super(MultiHeadAttention, self).__init__()
        self.supports_masking = True
        self.hidden_size = hidden_size
        self.num_heads = num_heads
        self.projection_dim = hidden_size // num_heads
        self.Q = layers.Dense(hidden_size)
        self.K = layers.Dense(hidden_size)
        self.V = layers.Dense(hidden_size)
        self.out = layers.Dense(hidden_size)

    def attention(self, query, key, value, mask):

        score = tf.matmul(query, key, transpose_b=True)
        dim_key = tf.cast(tf.shape(key)[-1], tf.float32)
        scaled_score = score / tf.math.sqrt(dim_key)
        max_length = tf.cast(tf.shape(scaled_score)[-1], tf.int64)
        m = tf.repeat(mask, max_length, axis=2) * (-1e9)
        scaled_score += m
        weights = tf.nn.softmax(scaled_score, axis=-1)
        output = tf.matmul(weights, value)
        return output, weights

    def separate_heads(self, x, batch_size):
        x = tf.reshape(x, (batch_size, -1, self.num_heads, self.projection_dim))
        return tf.transpose(x, perm=[0, 2, 1, 3])

    def call(self, inputs, att_mask):
        batch_size = tf.shape(inputs)[0]
        query = self.separate_heads(self.Q(inputs)  , batch_size)
        key = self.separate_heads(self.K(inputs), batch_size)
        value = self.separate_heads(self.V(inputs) , batch_size)
        attention, self.att_weights = self.attention(query, key, value, att_mask)
        attention = tf.transpose(attention, perm=[0, 2, 1, 3])
        concat_attention = tf.reshape(attention, (batch_size, -1, self.hidden_size))
        output = self.out(concat_attention)
        return output

    def compute_mask(self, inputs, mask=None):
        return mask  # Propagate the mask
```

#### Feed-Forward Sub-Layer

The feed-forward sub-layer in the encoder consists of two dense layers: an intermediate layer and an output layer, which projects back to the hidden layer size. Dropout is applied to the output of the intermediate layer. Unlike the original Transformer, BERT employs the GELU activation function in the intermediate dense layer. Since TensorFlow does not natively support [GELU](https://arxiv.org/abs/1606.08415 ) , you will need to implement it manually, by referencing the original paper.

```python
@tf.function

def GELU(x):

  ### YOUR CODE STARTS HERE ###
  return x*0.5*(1.0+ tf.math.erf(x/tf.math.sqrt(2.0))) # there are other approx but I'll stick to the first one

  ### YOUR CODE ENDS HERE ###
```

```python
class FFN(layers.Layer):

    def __init__(self, intermediate_size, hidden_size, drop_rate):
    
        super(FFN, self).__init__()
        self.intermediate = layers.Dense(intermediate_size, activation=GELU, kernel_initializer=TruncatedNormal(stddev=0.02))
        self.out = layers.Dense(hidden_size, kernel_initializer=TruncatedNormal(stddev=0.02))
        self.drop = layers.Dropout(drop_rate)
        
    def call(self, inputs):
    
        ### YOUR CODE STARTS HERE ###
        # I am getting the first output which comes from the intermediate layer
        output_interm=self.intermediate(inputs)
        # applying droupout to the output of intermediate layer
        output_dropout=self.drop(output_interm)
        # the output we received is now passed in the final layer
        output=self.out(output_dropout)
        return output
        ### YOUR CODE ENDS HERE ###
    
    def compute_mask(self,inputs,mask=None):
        return mask
```

#### Residual Connections

In the encoder, each sub-layer's output undergoes dropout, then is added to its input through a residual connection. This combined result then passes through layer normalization, creating an **AddNorm** layer that helps stabilize and improve the neural network's training process.

```python
class AddNorm(layers.Layer):

  def __init__(self, LNepsilon, drop_rate):

    super(AddNorm, self).__init__()
    self.supports_masking = True
    self.LN = layers.LayerNormalization(epsilon=LNepsilon)
    self.dropout = layers.Dropout(drop_rate)

  def compute_mask(self, inputs, mask=None):
    return mask

  def call(self, sub_layer_in, sub_layer_out):
    ### YOUR CODE STARTS HERE ###
    # 1- I'll apply the dropout of the sublayer output and add to input
    sub_output_drop=self.dropout(sub_layer_out)
    sub_output=sub_output_drop + sub_layer_in
    # 2- then pass through layer normalization
    output=self.LN(sub_output)
    return output
    ### YOUR CODE ENDS HERE ###
```

With all the essential components prepared, we're now ready to construct a complete encoder layer for the Transformer architecture.

```python
class Encoder(layers.Layer):

    def __init__(self, hidden_size, num_heads, intermediate_size, drop_rate=0.1, LNepsilon=1e-12):

      super(Encoder, self).__init__()
      self.attention = MultiHeadAttention(hidden_size, num_heads)
      self.ffn = FFN(intermediate_size, hidden_size, drop_rate)
      self.addnorm1 = AddNorm(LNepsilon, drop_rate)
      self.addnorm2 = AddNorm(LNepsilon, drop_rate)


    def call(self, inputs, mask):
      ### YOUR CODE STARTS HERE ###
      ## 1- I will pass the input to attnetion
      att_out=self.attention(inputs,mask)
      #2- apply resid+norm
      att_norm1= self.addnorm1(inputs,att_out)
      #3- apply feed forward nn
      ffn_out= self.ffn(att_norm1)
      #4- resid+norm
      ffn_norm2= self.addnorm2(att_norm1,ffn_out) # every time we add to the input
      return ffn_norm2
      ### YOUR CODE ENDS HERE ###

    def compute_mask(self, x, mask):
      return mask
```

In the next phase of BERT implementation, we'll focus on the embedding layer. For this single-sentence classification task, we'll omit segment embeddings and concentrate on creating token and positional embeddings.

Segment embedding in BERT is used to differentiate segments of a sentence. This identifies the segments each token belongs to, enabling the model to understand the structure of the input sentence. This is important for next sentence prediction or NLI where understanding the relationship between sentences and not only words is required.

```python

class Embedding(layers.Layer):
    def __init__(self, vocab_size, max_length, hidden_size):
        super(Embedding, self).__init__()
        self.TokEmb = layers.Embedding(input_dim=vocab_size, output_dim=hidden_size, mask_zero=True)
        self.PosEmb = tf.Variable(tf.random.truncated_normal(shape=(max_length, hidden_size), stddev=0.02))
        self.LN = layers.LayerNormalization(epsilon=1e-12)
        self.dropout = layers.Dropout(0.1)

    def build(self, input_shape):
        self.mask_layer = layers.Lambda(lambda x:
            1.0 - tf.cast(tf.not_equal(x, 0), tf.float32))
        super().build(input_shape)

    def call(self, inputs):
      ### YOUR CODE STARTS HERE ###
        # 1- get the positional embedding of the word
          seqlen=tf.shape(inputs)[1]
          positional_emb = self.PosEmb[:seqlen,:]
          # 2- get the token embedding so that we add them later on
          token_emb = self.TokEmb(inputs)
          #3- add the 2 together
          embedding=token_emb+positional_emb
          ## 4- now we apply the norm and dropout layers
          embedding_ln=self.LN(embedding)
          embedding_drop=self.dropout(embedding_ln)
          return embedding_drop
          ### YOUR CODE ENDS HERE ###

    def compute_mask(self, x, mask=None):
        mask = self.mask_layer(x)
        return mask[:, tf.newaxis, tf.newaxis, :]

```

The final layer to implement is the “pooler.” The pooler’s role is to transform the hidden states from the last encoder layer, with a shape of ￼, into a vector representation of shape ￼ for each input sentence. It achieves this by extracting the hidden state of the first token (a special token at the beginning of each sentence) and passing it through a dense layer with a tanh activation function, as used in the original implementation.

```python
class Pooler(layers.Layer):

    def __init__(self, hidden_size):

      super(Pooler, self).__init__()
      self.supports_masking = True
      self.dense = layers.Dense(hidden_size, activation='tanh')

    def call(self, encoder_out):

      ### YOUR CODE STARTS HERE ###
      x=encoder_out[:,0]
      pool=self.dense(x)
      return pool
      ### YOUR CODE ENDS HERE ###

    def compute_mask(self, inputs, mask=None):
        return None
```

```python
def BERT(vocab_size, max_length, hidden_size, num_layers, num_att_heads, intermediate_size, drop_rate=0.1):

  """
  Creates a BERT model based on the provided parameters.

  Arguments:
  - vocab_size: The size of the vocabulary.
  - max_length: The maximum length of each input sentence.
  - hidden_size: The dimensionality of the hidden states in each encoder layer.
  - num_layers: The number of encoder layers in the model.
  - num_att_heads: The number of attention heads in the multi-head attention mechanism.
  - intermediate_size: The dimensionality of the intermediate layer in the feed-forward sublayer of the encoders.
  - drop_rate: The dropout rate applied to all dropout layers in the model.

  Returns:
  - model: A BERT model with the specified architecture.
  """

  ### YOUR CODE STARTS HERE ###
  # inputs=tf.keras.layers.Input(shape=(max_length,), dtype=tf.int32) ## this is the input of the model
  inputs=keras.Input(shape=(max_length,),dtype=tf.int32)
  embedding_input=Embedding(vocab_size, max_length, hidden_size) ## 1st step to define the embedding inp
  embedding_output=embedding_input(inputs)
  mask = embedding_input.compute_mask(inputs) ##embed our data
  # mask = MyMask()(inputs)
  # mask = layers.Lambda(lambda x: tf.cast(tf.not_equal(x, 0), dtype=tf.float32))(inputs)
  ### nOW I can encode my data using the class Encoder defined above. Since we have multiple layer we will iterate through them to compute the output at every one
  for _ in range(num_layers):
      encoder=Encoder(hidden_size=hidden_size,
                               num_heads=num_att_heads,
                               intermediate_size=intermediate_size,
                               drop_rate=drop_rate)
      embedding_output=encoder(embedding_output, mask)
  pooler= Pooler(hidden_size)
  poolout=pooler(embedding_output)
  model = tf.keras.Model(inputs=inputs, outputs=[embedding_output, poolout])

  ### YOUR CODE ENDS HERE ###
  return model



```

```python
def convert_to_tokenss(s_list, max_length):
  toks, masks = [], []
  for x in s_list:
    enc = bert_tokenizer.encode_plus(x, max_length=max_length, pad_to_max_length=True)
    toks.append(enc['input_ids'])
    masks.append(enc['attention_mask'])
  return np.array(toks).astype(np.int64), np.array(masks).astype(np.int64)
```
